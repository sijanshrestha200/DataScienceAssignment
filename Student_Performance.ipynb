# Hypothesis Testing

Does Study Hours affect Test Score?

Hypotheses:

H₀: No relationship between Study Hours and Test Score

H₁: There is a relationship between study Hours and Test Score

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(df['Study_Hours'], df['Test_Score'])

print("Correlation:", corr)
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject H0 → Significant relationship exists")
else:
    print("Fail to reject H0 → No significant relationship")
```

# Conslusion
Since, p-value < 0.05, so,  we reject the Null Hypothesis (H₀) and accept the Alternative Hypothesis (H₁). i.e  there is a statistically significant positive relationship between Study_Hours and Test_Score.
