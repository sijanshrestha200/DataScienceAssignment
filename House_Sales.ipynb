# IQR (Inter Quartile Range)

```python
columns = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors','sqft_basement', 'sqft_above', 'long', 'grade']

for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    print(f"{col} IQR:", IQR)
```

# Hypothesis test
Does Living Area (sqft_living) affect house price?

Null Hypothesis (H0) : There is no significant relationship between sq_ft living and price

Alternative Hypothesis (H1) : There is a significant relationship between sq_ft living and price

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(df['sqft_living'], df['price'])

print("Correlation:", corr)
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject H0 → Significant relationship exists")
else:
    print("Fail to reject H0 → No significant relationship")
```

# Interpretation

The correlation test was used to determine the relationship between house size (sqft_living) and price. The p-value indicates whether the relationship is statistically significant

since p < 0.05 → strong evidence that house size affects price


# Conclusion 
The hypothesis test shows that there is a significant positive relationship between house size and price, meaning larger houses tend to have higher prices.
